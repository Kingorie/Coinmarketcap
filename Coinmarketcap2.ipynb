```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Webdriver to automate browser
driver = webdriver.Chrome()
driver.get('https://coinmarketcap.com/historical/20221009/')
# maximize window position
driver.maximize_window
```

```python
# Name
# Marketcap
# Price
# 24h_Change
# 7d%
# Vol_24h

Name_ = driver.find_elements("xpath", '//tbody/tr/td[2]/div//a[2]')
Marketcap_ = driver.find_elements("xpath", '//tbody/tr/td[4]/div')
Price_ = driver.find_elements("xpath", '//tbody/tr/td[5]')
Change_24h_ = driver.find_elements("xpath", '//tbody/tr/td[7]')
Change_7d_ = driver.find_elements("xpath", '//tbody/tr/td[10]')
Vol_24h_ = driver.find_elements(By.XPATH, "//tbody/tr/td[7]")

"""

# To confirm if code works:
print(len(Name)) #Number of cryptos

for n in Name:
     print(n.text)
     
  """  
```

```python
# Empty lists:


result = []

for i in range (len(Name_)):
    temporary_data = {'Crypto Name': Name_[i].text,
                      'Marketcap': Marketcap_[i].text,
                      'Price': Price_[i].text,
                      '% 24H': Change_24h_[i].text,
                      '% 7D': Change_7d_[i].text,
                      'Vol_24h': Vol_24h_[i].text}
    result.append(temporary_data)
```

```python
result_df = pd.DataFrame(result)
result
```
